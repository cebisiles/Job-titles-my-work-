```python
import numpy  as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/cebisiles/Job-titles-my-work-/main/joblisting_df.csv', index_col = False)
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
```

```python
df['title'] = df['title'].str.lower()
df['title'].head()
```

```python
titles_counts = df['title'].value_counts()
```

```python
titles_counts.head(10) 
```

```python
ax = titles_counts.head(10).plot.bar(rot=90) #Plotting the top 10 titles
```

```python
company_counts = df['company'].value_counts()
company_counts.head()
```

```python
ax = company_counts.head(10).plot.bar(rot=90)
```

```python
substring = 'remote'
contain_remote = df[df.apply(lambda row: row.astype(str).str.contains(substring, case=False).any(), axis=1)]
```

```python
contain_remote.head()
```

```python
company_counts = contain_remote['company'].value_counts()
ax = company_counts.head(10).plot.bar(rot=90)
```

```python
remote_title_counts = contain_remote['title'].value_counts()
ax = remote_title_counts.head(10).plot.bar(rot=90)
```
